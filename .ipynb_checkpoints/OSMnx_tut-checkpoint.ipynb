```python
import osmnx as ox
import networkx as nx

ox.__version__
```

### Working with street network

```python
city = ox.geocode_to_gdf("London, UK") # 获取 London, UK的边界轮廓
city = city.to_crs(epsg=3857)

# 绘图
ax = city.plot()
ax.axis("off")
```

```python
# 返回的是 查询城市的graph对象，包含node和edge
G = ox.graph_from_place("City of Westminster, London, UK", network_type="bike")
fig, ax = ox.plot.plot_graph(G, bgcolor="white", node_color="red", edge_color="green", edge_linewidth=0.5, node_size=2)
print(G)
```

```python
# get a fully bidirection(双向) network (as a MultiDiGraph)
# OSMnx 默认只有 ‘drive’ 是自动处理为双向网络的，所以 “把 bike 网络也当作双向的来处理。” 否则某些单向道在骑行图中就只能走一边。
ox.settings.bidirectional_network_types += "bike"
G = ox.graph_from_place("City of Westminster, London, UK", network_type="bike")

# convert your MultiDiGraph to an undirected MultiGraph
M = ox.convert.to_undirected(G)

# convert your MultiDiGraph to a DiGraph without parallel edges
D = ox.convert.to_digraph(G)
```

```python
# you can convert your graph to node and edge GeoPandas GeoDataFrames
gdf_nodes, gdf_edges = ox.convert.graph_to_gdfs(G)
gdf_nodes.head()
```

```python
gdf_edges.head()
```

```python
# convert node/edge GeoPandas GeoDataFrames to a NetworkX MultiDiGraph
G2 = ox.convert.graph_from_gdfs(gdf_nodes, gdf_edges, graph_attrs=G.graph)
```

### Basic street network stats

```python
# what sized area does our network cover in square meters?
G_proj = ox.projection.project_graph(G)
nodes_proj = ox.convert.graph_to_gdfs(G_proj, edges=False) # 将图中的节点转为 GeoDataFrame，每个节点都有 geometry（点坐标）
graph_area_m = nodes_proj.union_all().convex_hull.area
print(f"道路网络的覆盖面积：{graph_area_m} 平方米")
```

```python
# show some basic stats about the network
ox.stats.basic_stats(G_proj, area=graph_area_m, clean_int_tol=15)
```

### Visualize street centrality

```python
edge_centrality = nx.closeness_centrality(nx.line_graph(G))
nx.set_edge_attributes(G, edge_centrality, "edge_centrality")
```

```python
# color edges in original graph with closeness centralities from line graph
ec = ox.plot.get_edge_colors_by_attr(G, "edge_centrality", cmap="inferno")
fig, ax = ox.plot.plot_graph(G, node_color="green", edge_color=ec, edge_linewidth=2, node_size=0)
```

### Routing

```python
# 为每条边添加 speed_kph 属性，如果没有提供限速数据，就用默认值填充（基于道路类型）
G = ox.routing.add_edge_speeds(G)
print(G)
```

```python
# 使用边的长度和限速，计算每条边的旅行时间（秒）
G = ox.routing.add_edge_travel_times(G)
print(G)
```

```python
# get the nearest network nodes to two lat/lng points with the distance module
orig = ox.distance.nearest_nodes(G, X=51.523627, Y=-0.152799)
dest = ox.distance.nearest_nodes(G, X=30.534512, Y=0.164451)
```

```python
# find the shorest path between nodes, minizing travel time, then plot it
route = ox.routing.shortest_path(G, orig, dest, weight="travel_time")
fig, ax = ox.plot.plot_graph_route(G, route, node_size=0)
```

```python
# how long is our route in meters
edge_lengths = ox.routing.route_to_gdf(G, route)["length"]
print(f"The total distance is {round(sum(edge_lengths))} m")
```

```python
# 计算两点之间的zhi'xi
orig_x = G.nodes[orig]['x']
orig_y = G.nodes[orig]['y']
dest_x = G.nodes[dest]['x']
dest_y = G.nodes[dest]['y']
round(ox.distance.great_circle(orig_y, orig_x, dest_y, dest_x))
```
